## Question 2: An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions.

A. Probability that none of the items are defective?

B. Probability that exactly one of the items is defective?

C. Probability that two or fewer of the items are defective?

D. Probability that three or more of the items are defective ?

```python
import numpy as np
import scipy.stats as stats

n=10 ## Number of samples
p=0.05 ## Failure rate
k=np.arange (0,11) ## Total 11 probabilities (including the first with 0)
k
```

```python
binomial = stats.binom.pmf(k,n,p)
```

```python
## Answer to A : Probability that none of the items are defective
Pa = binomial[0]
print ("Answer to Q1.A : Probability that none of the items are defective:", Pa)

## Answer to B : Probability that exactly one of the items is defective
Pb = binomial [1]
print ("Answer to Q1.B : Probability that exactly one of the items are defective:", Pb)

## Answer to C : Probability that two or fewer of the items are defective
Pc = binomial[0]+binomial[1]+binomial[2]
print ("Answer to Q1.C : Probability that two or fewer of the items are defective:", Pc)

## Answer to D : Probability that three or more of the items are defective
Pd = 1 - Pc
print ("Answer to Q1.D : Probability that two or fewer of the items are defective:", Pd)
```

## Question 3: A car salesman sells on an average 3 cars per week.

A. Probability that in a given week he will sell some cars.

B. Probability that in a given week he will sell 2 or more but less than 5 cars.

C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold perweek.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

#average rate, denoted with lambda in the formula for poisson distribution 

rate = 3

#numpy array of different no. of successes ranging from 0 to 50 to construct a probability distribution

n=np.arange(0,50)

#calculating the distribution and storing the distribution of probablitites in an array
poisson = stats.poisson.pmf(n,rate)

# ANS A: Probability that in a given week he will sell some cars. 1 - Probability of selling 0 cars
Pa = 1 - poisson[0]
print("Answer for Q3:A: Probability that in a given week he will sell some cars: ", Pa)

# ANS B: Probability that in a given week he will sell 2 or more but less than 5 cars
# This is the sum of probabilities for selling 2, 3 and 4 cars per week
Pb = poisson[2]+poisson[3]+ poisson[4]
print("Answer for Q3:B: Probability that in a given week he will sell 2 or more but less than 5 cars: ", Pb)




```

```python
#Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold perweek.
plt.plot(n,poisson,'o-')
plt.title('Poisson: $\lambda$ = %i ' % rate)
plt.xlabel('Number of Cars Sold per week')
plt.ylabel('Probability of Cars sold per week')
plt.show()
```

## Question 4: Accuracy in understanding orders for a speech based bot at a restaurant is important for the Company X which has designed, marketed and launched the product for a contactless delivery due to the COVID-19 pandemic. Recognition accuracy that measures the percentage of orders that are taken correctly is 86.8%. Suppose that you place order with the bot and two friends of yours independently place orders with the same bot. Answer the following questions.

A. What is the probability that all three orders will be recognised correctly?

B. What is the probability that none of the three orders will be recognised correctly?

C. What is the probability that at least two of the three orders will be recognised correctly?

```python
# This is an example of Binomial distribution

import numpy as np
import scipy.stats as stats

n=3 ## Number of samples
p=0.86 ## Success rate
k=np.arange (0,4)

binomial = stats.binom.pmf(k,n,p)

#A. What is the probability that all three orders will be recognised correctly?
Pa = binomial[3]
print("Ans Q3A: Probability of All three orders being recognised correctly: ", Pa)

#B. What is the probability that none of the three orders will be recognised correctly?
Pb = binomial[0]
print("Ans Q3B: Probability of none of the orders being recognised correctly: ", Pb)

#C. What is the probability that at least two of the three orders will be recognised correctly?
Pc = binomial[2] + binomial[3]
print("Ans Q3C: Probability of at least two orders being recognised correctly: ", Pc)
```

## Question 5: A group of 300 professionals sat for a competitive exam. The results show the information of marks obtained by them have a mean of 60 and a standard deviation of 12. The pattern of marks follows a normal distribution. Answer the following questions.

A. What is the percentage of students who score more than 80.

B. What is the percentage of students who score less than 50.

C. What should be the distinction mark if the highest 10% of students are to be awarded distinction?

```python
import numpy as np
import scipy.stats as stats

#ANS for part A:

# This is the normal distribution with mean as 60 and standard deviation as 12.
# To calculate the probability of population more than 80, calculate the probability to the left. 
# Subtract that figure from 1 to get the probability for the value to the right of 80.
Pa = 1 - stats.norm.cdf (80,loc=60, scale=12)

print("ANS Q5A: percentage of students who score more than 80:", Pa*100)



#ANS for part B:

Pb = stats.norm.cdf (50,loc=60, scale=12)
print("ANS Q5B: percentage of students who score less than 50:", Pb*100)

#ANS for Part C:
# Use norm.ppf function to identify the value on curve for 90% on the curve

DistMark = stats.norm.ppf (0.90, 60, 12)

print("ANS Q5C: Distinction Mark for upper 10% students", DistMark)
```

```python
## All the currency values are of type string based on above output
## Need to get rid of $ at the beginning of the string. Write a function for the same
def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)

ccdatafmt['Funding'] = ccdatafmt['Funding'].apply(clean_currency)
```

```python
## All the currency values are of type string
## Need to convert into numerical by removing K and M by applying right multification factor

## This function takes a value as x and if it is string, it will convert to float by multiplying with 
## right factor (1000 for K, 1000000 for M and so on)
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0

## Apply function to all values of column 'Funding'
ccdatafmt['Funding'] = ccdatafmt['Funding'].apply(value_to_float)

ccdatafmt.tail(25)
```

### Null and Alternate Hypothesis
##### Null Hypothesis
Funding position for Operating and Closed companies follow the similar distribution

#### Alternae Hypothesis
Funding position for Operating and Closed companies have different distributions

#### Significance Level
Here we select  = 0.05

#### Test Statistics
We are having a categorical variable (OperatingState) and one continuous variable (Funding) and we want to validate curve for two different values of the categrical variable. This qualifies for Innova test

```python
import statsmodels.api         as     sm
from   statsmodels.formula.api import ols
 
mod = ols('Funding ~ OperatingState', data = ccdatafmt_Operating_Closed).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print(aov_table)
```

Here the P value is 0.415198 which is far higher than our level of Significance (0.05)
So we fail to reject the hypothesis.

The inference is that from the data provided the funding position for both closed and Operating companies do not have significance difference

```python
## No of winners who are operating
nOperating_Winners = ccdatafmt_Operating[ccdatafmt_Operating['Result'] == 'Winner'].Result.value_counts()[0]  
## No of Contestants who are operating
nOperating_Contestants = ccdatafmt_Operating[ccdatafmt_Operating['Result'] == 'Contestant'].Result.value_counts()[0] 
## No of total winners
nTotal_Winners = ccdatafmtorig[ccdatafmtorig['Result'] == 'Winner'].Result.value_counts()[0]
## No of total Contestants
nTotal_Contestants = ccdatafmtorig[ccdatafmtorig['Result'] == 'Contestant'].Result.value_counts()[0]

print ('Total winners who are operating:', nOperating_Winners)

print ('Total contestants who are operating:', nOperating_Contestants)
print ('Total winners:', nTotal_Winners)
print ('Total contestants:', nTotal_Contestants)
```

##### Test of proportion needs to be applied here:

```python
from statsmodels.stats.proportion import proportions_ztest

stat, pval = proportions_ztest([nOperating_Winners, nOperating_Contestants] , [nTotal_Winners, nTotal_Contestants])

print ('pval is:', pval)

if pval < 0.05:
    print(f'With a p-value of {round(pval,4)} the difference is significant. Null hypothesis is rejected')
else:
    print(f'With a p-value of {round(pval,4)} the difference is not significant. We fail to reject null hypothesis')
```
